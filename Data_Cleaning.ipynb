```python
import pandas as pd
```

```python
df=pd.read_csv('glassdoor_jobs.csv')
```

```python
df.head(5)
```

```python
df.shape
```

```python
df = df[df['Salary Estimate'] != '-1']
```

```python
df.shape
```

LAMBDA FUNCTION to parse

### SALARY PARSING

```python
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
salary
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(salary)
```

```python
minuskd = salary.apply(lambda x: x.replace('K','').replace('$',''))
```

```python
minuskd
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(minuskd)
```

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['employer provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
```

```python
min_hr = minuskd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(min_hr)
```

```python
df['min_salary'] = min_hr.apply(lambda x: int(x.split('-')[0]))
df['max_salary'] = min_hr.apply(lambda x: int(x.split('-')[1]))
df['avg_salary'] = (df.min_salary+df.max_salary)/2
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df)
```

### COMPANY NAME

```python
df['company_text'] = df.apply(lambda x: x['Company Name'] if x['Rating']<0 else x['Company Name'][:-3], axis=1)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df)
```

### STATE FIELD

```python
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
```

```python
df.job_state.value_counts()
```

```python
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
```

### AGE OF COMPANY

```python
df['age'] = df.Founded.apply(lambda x: x if x<1 else 2021-x)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df)
```

### PARSE JOB DESCRIPTIONS

```python
df['Job Description'][1]
```

```python
#python
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
#r studio
df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() else 0)
#spark
df['spark_yn'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
#aws
df['aws_yn'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
#excel
df['excel_yn'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df)
```

```python
df_out = df.drop(['Unnamed: 0'], axis=1)
```

```python
df_out.to_csv('salary_data_cleaned.csv',index=False)
```
